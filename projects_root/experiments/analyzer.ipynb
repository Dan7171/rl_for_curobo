```python
with open(path, 'rb') as f:
    stats = pickle.load(f)

task_statman_name, task_stats = 'task_stats', stats['task_stats']
a_statman_names, agents_stats = [statman_name for statman_name in stats.keys() if 'agent' in statman_name], [stats[statman_name] for statman_name in stats.keys() if 'agent' in statman_name]
all_stats = [task_stats] + agents_stats


def print_statman_names(stats):
    print('-'*100)
    print(f'N statmanagers: {len(stats)}')
    for statman_name in stats:
        print(f'statman name: {statman_name}')
    print('-'*100)

def print_stat_names(stats):
    
    for statman_name in stats:
        print('-'*100)
        print(f'statman name: {statman_name}')
        stat_names = stats[statman_name].keys()
        for stat_name in stat_names:
            print(f'    stat name: {stat_name}')
        print('-'*100)

def print_stat_counts(stats):
    for statman_name in stats:
        print('-'*100)
        print(f'statman name: {statman_name}')
        stat_names = stats[statman_name].keys()
        for stat_name in stat_names:
            print(f'    stat name: {stat_name}')
            print(f'    N records: {len(stats[statman_name][stat_name])}')
        print('-'*100)

def print_stat_counts_with_examples(stats):
    for statman_name in stats:
        print('-'*100)
        space = '    '
        print(f'statman name: {statman_name}')
        stat_names = stats[statman_name].keys()
        ind2 = f'{space}{space}'
        ind3 = f'{ind2}{space}'
        for stat_name in stat_names:
            print(f'{ind2}stat name: {stat_name}')
            n_records = len(stats[statman_name][stat_name])
            print(f'{ind2}N records: {n_records}')
            example_indices = [0,n_records//2,n_records-1]
            examples_headers = ['first','middle','last']
            # print(f'{space}{space}examples:{example_indices}')
            for i in range(len(example_indices)):
                example_idx = example_indices[i]
                stat_keys, stat_vals = stats[statman_name][stat_name][example_idx]                
                print(f'{ind2}{examples_headers[i]}({example_idx}):')
                print(f'{ind3}K:{stat_keys}')
                print(f'{ind3}V:{stat_vals}')
        
        print('-'*100)

def info(stats,verbosity=1):

    if verbosity == 1:
        print_statman_names(stats)
    if verbosity == 2:
        print_statman_names(stats)
        print_stat_names(stats)
    if verbosity == 3:
        print_stat_counts(stats)
    if verbosity == 4:
        print_stat_counts_with_examples(stats)


info(stats,verbosity=4)



```
